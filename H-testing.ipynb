## Тест Вальда

Тест Вальда не редко называется тестом значимости. Основополагающим принципимом его работы служит оценка регрессии без ограничений с последующей оценкой того, совместимы ли результаты с гипотезой в пределазх выборочной вариации. 

Самый простой случай - проверка равенства какого-либо коэффициента другому коэффициенту. 
$H_0: \beta_2=\beta^{0}_2$ 
$\beta^{0}_2$ - гипотетическое значение коэффициента. 

Расстояние Вальда - линейное расстояние между оценкой коэффициента и гипотетическим значением, разница между этими значениями измеренное в единицах стандартных отелонений

$W_k=\frac{b_k-\beta^{0}_k}{\sqrt{\sigma^2 S^{kk}}}$
